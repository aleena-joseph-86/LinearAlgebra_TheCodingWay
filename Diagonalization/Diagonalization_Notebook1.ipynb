#### What does T mean here?
Here T is a linear transformation between R<sup>2</sup> and  R<sup>2</sup>.</br>
A `linear transformation` (or a linear map) is a function T:R<sup>n</sup>R<sup>m</sup>
that satisfies the following properties:

                            1.T(x+y)=T(x)+T(y)
                            2.T(ax)=aT(x)                            
for any vectors x,yR<sup>n</sup> and any scalar aR.

#### What does basis vectors mean?</br>
In a vector space, if a set of vectors can be used to express every vector in the space as a unique linear combination of those vectors, and those vectors are linearly independent (meaning that none of them can be expressed as a linear combination of the others), then we refer to them as `basis vectors` for that vector space.

#### 1.Import the necessary libraries

```python
import numpy as np 
import matplotlib.pyplot as plt
```

#### 2.Define the necessary matrix and the vectors

```python
#  Define the diagonal matrix
D = np.array([[3,0],
              [0,6]])

# Define the basis vectors
e1 = np.array([[1],
               [0]])
e2 = np.array([[0],
               [1]])

# Define the vector x 
x = np.array([[1],
              [1]])

# Diplay D,e1,e2 and x
print('Matrix D\n',D)
print('\nBasis vector e1 \n',e1)
print('\nBasis vector e2 \n',e2)
print('\nvector x \n',x)
```

#### 3.Perform the necessary transformations

```python
# Transform e1,e2 and x using the Diagonal Matrix
De1 = np.matmul(D,e1)
De2 = np.matmul(D,e2)
Dx = np.matmul(D,x)

# Display De1,De2 and Dx
print('\nDe1 \n',De1)
print('\nDe2 \n',De2)
print('\nDx \n',Dx)
```

#### 4.Plot the transformations

```python
fig, (ax) = plt.subplots()

# Plot e1,e2 and x
ax.arrow(0,0,e1[0][0],e1[1][0],head_width=0.2,head_length=0.2,facecolor='red',edgecolor='red') #e1
ax.arrow(0,0,e2[0][0],e2[1][0],head_width=0.2,head_length=0.2,facecolor='blue',edgecolor='blue') #e2
ax.arrow(0,0,x[0][0],x[1][0],head_width=0.2,head_length=0.2,facecolor='yellow',edgecolor='yellow') #x

# Plot the Transformations De1,De2,Dx
ax.arrow(0,0,De1[0][0],De1[1][0],head_width=0.2,head_length=0.2,facecolor='red',edgecolor='red') #De1
ax.arrow(0,0,De2[0][0],De2[1][0],head_width=0.2,head_length=0.2,facecolor='blue',edgecolor='blue') #De2
ax.arrow(0,0,Dx[0][0],Dx[1][0],head_width=0.2,head_length=0.2,facecolor='yellow',edgecolor='yellow') #Dx

# Annotate the plots 
ax.annotate('e1',xy=(e1[0][0],e1[1][0]),xytext=(e1[0][0]+.1,e1[1][0]+.1))
ax.annotate('e2',xy=(e2[0][0],e2[1][0]),xytext=(e2[0][0]+.1,e2[1][0]+.1))
ax.annotate('x',xy=(x[0][0],x[1][0]),xytext=(x[0][0]+.1,x[1][0]+.1))
ax.annotate('De1',xy=(De1[0][0],De1[1][0]),xytext=(De1[0][0]+.1,De1[1][0]+.1))
ax.annotate('De2',xy=(De2[0][0],De2[1][0]),xytext=(De2[0][0]+.1,De2[1][0]+.1))
ax.annotate('Dx',xy=(Dx[0][0],Dx[1][0]),xytext=(Dx[0][0]+.1,Dx[1][0]+.1))

# Display the graph 
plt.grid()
plt.show()
```

#### Observations
* On transforming e1 and e2 with D, e1 gets scaled on x axis and e2 gets scaled on y axis.
* On transforming x with D, x gets scaled differently across x and y axis

#### 5.Calculate D<sup>2</sup> and D<sup>3</sup>

```python
# Normal matrix multiplication
D_square_normal_matrix_mul = D@D
print('D_square\n',D_square_normal_matrix_mul)
D_cube_normal_matrix_mul =  D@D@D
print('D_cube\n',D_cube_normal_matrix_mul)
```

#### Observations
* When a diagonal matrix is multiplied by itself n times, each diagonal element gets raised to the power of .
* Squaring or cubing a diagonal matrix does not change the order of its diagonal elements but affects their magnitudes based on their original signs (positive or negative).
