1.Why should A be a N * N matrix?Why not a M * N matrix?

        An eigenvector x is a non-zero vector that, when multiplied by the matrix A, results in a scaled version of itself. This relationship is represented as Ax = λx, where λ is the corresponding eigenvalue.

In a rectangular matrix, such as m × n where m ≠ n, the operation Ax does not result in a scaled version of x because the dimensions are incompatible.

Proof:

    (N * N) x (N * 1) = N * 1 (scaled version of itself)
    (M * N) x (N * 1) = M * 1 (not a scaled version of itself)
  
Conclusion:

    Hence A must be a N * N matrix
    
2.What does non-trivial in the definition of eigenvector mean?

        

# Problem Statement

                    let A=[1 6   u=[6 -5] and v=[3 -2]
                           5 2]   
                    Are u and v eigenvectors of A? Prove that visually.
                    
                    
Solution:

        To determine whether u and v are eigenvectors of matrix A, we need to check if the following equations hold:

                                           A.u=λu.u
                                           A.v=λv.v

 where λu and λv are the eigenvalues corresponding to u and v respectively.

```python
# install numPy and matplotlib using pip command by running cmd in administrator mode
# import numPy and matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Declare and initialize the matrix A
A = np.array([[1, 6],
              [5, 2]])

# Declare and initialize  vectors u and v
u = np.array([6, -5])
v = np.array([3, -2])

# Calculate A*u and A*v
Au = np.dot(A, u)
Av = np.dot(A, v)

# Calculate eigenvalues
# Concept used is EVD(eigenvalue decomposition i.e A=PDP^-1) where P->eigenvectors and D->diagonal matrix of eigenvalues of A
# np.linalg.eig(A) returns P as a matrix and D as a vector
eigenvalues, eigenvectors = np.linalg.eig(A) 

# Calculate lambda1*u and lambda2*v
lambda1_u = eigenvalues[0] * u
lambda2_v = eigenvalues[1] * v

# Set the space the plotting
plt.figure(figsize=(10, 5))

# Plot A*u, A*v, lambda1*u, lambda2*v 
plt.quiver([0, 0, 0, 0], [0, 0, 0, 0], [Au[0], Av[0], lambda1_u[0], lambda2_v[0]], [Au[1], Av[1], lambda1_u[1], lambda2_v[1]], color=['r', 'b', 'g', 'purple'], scale=30, linewidth=1.5)
plt.xlim(-100, 100)
plt.ylim(-100, 100)
plt.title('Vectors')

# Add labels next to each plot 
for label, x, y in zip(['A*u', 'A*v', 'λ*u', 'λ*v'], [Au[0], Av[0], lambda1_u[0], lambda2_v[0]], [Au[1], Av[1], lambda1_u[1], lambda2_v[1]]):
    plt.text(x, y, label, fontsize=10, ha='center', va='center')

# name the x and y axis
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# set up the graph sheet
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)

# used for ensuring proper spacing when subplots are used
plt.tight_layout()
plt.show()
```

            From the plot above,its clear that A.u is equal to λu.u and A.v not equal to λv.v.Therefore,u is an eigenvector 
    of A and v is not an eigenvector of A.
