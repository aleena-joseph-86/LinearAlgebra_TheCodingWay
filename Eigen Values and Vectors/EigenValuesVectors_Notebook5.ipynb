#### 1.Import the necessary libraries

```python
import numpy as np
```

#### 2.Initialize the probability matrix

```python
# Declare and initialize the transition_matrix
probability_matrix = np.array([
    [0,0,1,0.5],
    [0.333333,0,0,0],
    [0.333333,0.5,0,0.5],
    [0.333333,0.5,0,0]
])

# Initialize the no of iterations
no_of_iterations_given = 100 # from given

# store all the pages 
pages = ['page1','page2','page3','page3']
```

#### 3.Initial probability of landing on other pages from page 1

```python
# accessing column one of the probabililty_matrix
initial_probability_from_page1 = probability_matrix[:,0]

# display the initial probability of landing on other pages fron page 1
for page_no,initial_probability in zip(pages,initial_probability_from_page1):
    print(f'The initial probability of landing on {page_no} from page1 is {initial_probability}\n')
```

#### 4.Probability of landing on other pages from page 1 after the 100th iteration

```python
# assign the initial probability of reaching other pages from page 1 to pprobabilities
probabilities = initial_probability_from_page1[:]

for _ in range(no_of_iterations_given):
     """
     The probability of reaching other pages from a particular page after the ith iteration is given by
     
                             probability(i) = probability_matrix * probability(i-1)
     """
     probabilities = np.matmul(probability_matrix,probabilities)

# display the probability of reaching other pages from page 1 after ith iteration
for page_no,probability in zip(pages,probabilities):
    print(f'The probability of landing on {page_no} from page1 after 100 iterations is {round(probability,2)}\n')
```

#### 5.Initial probability of landing on other pages from page 3

```python
# accessing column three of the pprobabililty_matrix
initial_probability_from_page3 = probability_matrix[:,2]

# display the initial probability of landing on other pages fron page 3
for page_no,initial_probability in zip(pages,initial_probability_from_page3):
    print(f'The initial probability of landing on {page_no} from page3 is {initial_probability}\n')
```

#### 6.Probability of landing on other pages from page 3 after the 100th iteration

```python
# assign the initial probability of reaching other pages from page 3 to probabilities
probabilities = initial_probability_from_page3[:]

for _ in range(no_of_iterations_given):
    probabilities = np.matmul(probability_matrix,probabilities)
    
# display the probability of reaching other pages from page 3 after ith iteration
for page_no,probability in zip(pages,probabilities):
    print(f'The probability of landing on {page_no} from page3 after 100 iterations is {round(probability,2)}\n')
```

#### Conclusion
 From step 4 and step 6 it is clear that the probability of landing on other pages from a particular page converges to steady probability vector known as the `steady state vector`
